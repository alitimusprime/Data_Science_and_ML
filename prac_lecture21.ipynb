```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10,100)
y1 = np.sin(x)
plt.figure(figsize= (10,5))
plt.plot(x,y1)
plt.title("Line plot", color = "green" )
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10,5))
y2 = (np.tan(x))
plt.plot(x,y1, label = "Sin wave", color = "red")
plt.plot(x,y2, label = "Tan wave", color = "green")

plt.title("Sine and Tangent Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x, y1, color='green', marker='o')
plt.title("Scatter Plot of Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Combining Line and Scatter Plots
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Line (Sine)', color='blue', linestyle='-')
plt.scatter(x, y1, color='orange', marker='o', label='Scatter Points')
plt.title("Combined Line and Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
