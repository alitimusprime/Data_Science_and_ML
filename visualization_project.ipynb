```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cancer = pd.read_csv("cancer.csv")
```

```python
data = cancer
data
```

```python
data.columns
```

```python
data['Class'] = data['Class'].map({0: 'benign', 1: 'malignant'})
```

```python
# Plotting the distribution of sepal length for each species
plt.figure(figsize=(10, 6))
for i in data['Class'].unique():
    subset = data[data['Class'] == i]
    plt.hist(subset['menopause'], alpha=0.5, label=i)

plt.title('cancer through menopause')
plt.xlabel('menopause')
plt.ylabel('class')
plt.legend()
plt.grid(True)
plt.show()
```
