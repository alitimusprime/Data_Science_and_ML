***1.) Arithmetic Series***

```python
#maths formula
n = 100000000000
start = time.time()
sn = ((n/2) * ((2*(2) + (n-1)*2)))
print (sn)
print (time.time() - start)
```

```python
#loop method
start = time.time()
for n in range (1000000):
    n = n +2

print (n)
print (time.time() - start)
```

***2.)SUM OF SQUARE OF N NUMBERS***

```python
#maths formula
n = 100000000000
start = time.time()
sqn = ((n*(n+1)*(2*(n)+1)) / 6)
print (sqn)
print (time.time() - start)
```

```python
start = time.time()
for abc in range(1000**2):
    abc = (abc + (abc**2))


print (abc)
print (time.time() - start)
```

***3.)Diagonal Count in a Polygon***

```python
n = 1000000000
start = time.time()
pn = (n*(n-3)/2)
print (pn)
print (time.time() - start)
```

```python
n = 1000000
start = time.time()
diagonals = 0
for i in range(n - 3): 
    diagonals += (n - i - 3)
print(diagonals//2)
print (time.time() - start)
```

***6.) Sum Of Cube Of N Numbers***

```python
n = 100000
start = time.time()
sum_of_cubes = 0
for i in range(1, n + 1):
    sum_of_cubes += i**3
print(sum_of_cubes)
print (time.time() - start)
```

***7.)Sum Of Triangular Numbers***

```python
start = time.time()
n = 100000
triangular_sum = (n * (n + 1) * (n + 2)) // 6 
print(triangular_sum)
print (time.time() - start)
```

```python
start = time.time()
n = 100000
triangular_sum = 0

for i in range(1, n + 1):  
    number = ((i * (i + 1)) // 2)
    triangular_sum += number 

print(triangular_sum)
print(time.time()-start)
```

***8.)Count Of Multiples of K in Range N***

```python
#Formula Method
start = time.time()
n = 1000000
k = 5
count_multiples = n // k
print(count_multiples)
print(time.time() - start)
```

```python
#Loop Method
start = time.time()
n = 100000
k = 5
count_multiples = 0

for i in range(1, n + 1):
    if i % k == 0:
        count_multiples += 1

print(count_multiples)
print(time.time()-start)
```

***9.) Product of First N odd Numbers***

```python
#Formula Method
start = time.time()
n = 1000
product= math.factorial(2 * n) // (2 ** n * math.factorial(n))
print(product)
print(time.time() - start)
```

```python
#loop method
n = 1000
product = 1
start = time.time()
for i in range(1, 2 * n, 2):
    product *= i

print(product)
print(time.time()-start)
```

***10.) Sum Of Odd Numbers till N***

```python
#Formula Method
start = time.time()
n = 20000000
sum_odd = n ** 2
print( sum_odd)
print(time.time()- start)
```

```python
#Loop Method
n = 20000000
sum_odd = 0
current_odd = 1
start = time.time()

for i in range(n):
    sum_odd += current_odd
    current_odd += 2

print(sum_odd)
print(time.time() - start)
```
